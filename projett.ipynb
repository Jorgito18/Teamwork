```python
import pandas as pd
import xlrd
```

```python
fichier_artisans = 'artisans_etalab.xls'
nombre_de_lignes = 103 # 103 car on commence à partir de la 3e ligne
colonne_alimentation = 6
colonne_fabrication = 7
colonne_batiment = 8
colonne_services = 9

dt = pd.read_excel(fichier_artisans, sheetname = 'Comptages', header = 2)
s = xlrd.open_workbook(fichier_artisans)
dt

# tableaux pour les valeurs finales
valeursMax = []
valeursMin = []

# tableaux pour les valeurs intermédiaire
valeurs_alimentation = []
valeurs_fabrication = []
valeurs_batiment = []
valeurs_services = []

with open(fichier_artisans, 'r') as f:
    for ligne in range(3, nombre_de_lignes):
        valeur = s.sheets()[0].cell(ligne, colonne_alimentation).value
        valeurs_alimentation.append(valeur)

with open(fichier_artisans, 'r') as f:
    for ligne in range(3, nombre_de_lignes):
        valeur = s.sheets()[0].cell(ligne, colonne_fabrication).value
        valeurs_fabrication.append(valeur)
        
with open(fichier_artisans, 'r') as f:
    for ligne in range(3, nombre_de_lignes):
        valeur = s.sheets()[0].cell(ligne, colonne_batiment).value
        valeurs_batiment.append(valeur)

with open(fichier_artisans, 'r') as f:
    for ligne in range(3, nombre_de_lignes):
        valeur = s.sheets()[0].cell(ligne, colonne_services).value
        valeurs_services.append(valeur)
        
        
#print(valeurs_alimentation)
#print(valeurs_fabrication)
#print(valeurs_batiment)
#print(valeurs_services)

meilleur_alimentation = max(valeurs_alimentation)
meilleur_fabrication = max(valeurs_fabrication)
meilleur_batiment = max(valeurs_batiment)
meilleur_services = max(valeurs_services)

print(meilleur_alimentation)
print(meilleur_fabrication)
print(meilleur_batiment)
print(meilleur_services)

for i in range(nombre_de_lignes-3):
    if valeurs_alimentation[i] == meilleur_alimentation:
        index_meilleur_alimentation = i
    if valeurs_fabrication[i] == meilleur_fabrication:
        index_meilleur_fabrication = i
    if valeurs_batiment[i] == meilleur_batiment:
        index_meilleur_batiment = i
    if valeurs_services[i] == meilleur_services:
        index_meilleur_services = i
        
print(index_meilleur_alimentation)
print(index_meilleur_batiment)
print(index_meilleur_fabrication)
print(index_meilleur_services)
```
